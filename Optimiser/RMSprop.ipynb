Sigmoid activation function

$\sigma(x) = \frac{1}{1 + e^{-x}}$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

Standar Deviation

$SDT = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$

```python
def standardize(x):
  return (x - x.mean()) / x.std()
```

Log Loss Function

$L = - \frac{1}{n} \sum_{i=1}^{n}\left[ y_i \log(p_i) + (1 - y_i)\log(1 - p_i) \right]$

```python
def Log_loss(y_test,y_predicted):
  epsilon=1e-15
  y_predicted_new = np.maximum(y_predicted,epsilon)
  y_predicted_new = np.minimum(y_predicted_new,1-epsilon)
  return -np.mean(y_test*np.log(y_predicted_new)+(1-y_test)*np.log(1-y_predicted_new))
```

RMSprop Gradient Descent function

$w = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}_i - y_i)x_i$

$b = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}_i - y_i)$

$v_w = \beta \, v_w + (1 - \beta)\, w^2$

$v_b = \beta \, v_b + (1 - \beta)\, b^2$

$weight = weight - \frac{learning\ rate}{\sqrt{v_w} + \epsilon} \times w$

$bias = bias - \frac{learning\ rate}{\sqrt{v_b} + \epsilon} \times b$

```python
def RMSprop(X, y, epochs=1000, lr=0.01, beta=0.9):

    # Convert to numpy
    X = np.array(X)
    y = np.array(y)

    # Standardize (column-wise using your function)
    for i in range(X.shape[1]):
        X[:, i] = standardize(X[:, i])

    n_samples, n_features = X.shape

    # Initialize
    weights = np.zeros(n_features)
    bias = 0

    # RMSprop accumulators
    v_w = np.zeros(n_features)
    v_b = 0

    epsilon = 1e-8

    losses = []
    accuracies = []
    epoch_list = []

    for i in range(epochs):

        # Forward
        linear = np.dot(X, weights) + bias
        y_pred = sigmoid(linear)

        # Loss
        loss = Log_loss(y, y_pred)

        # Gradients
        error = y_pred - y
        dw = np.dot(X.T, error) / n_samples
        db = np.mean(error)

        # RMSprop update
        v_w = beta * v_w + (1 - beta) * (dw ** 2)
        v_b = beta * v_b + (1 - beta) * (db ** 2)

        # Parameter update
        weights -= (lr / (np.sqrt(v_w) + epsilon)) * dw
        bias -= (lr / (np.sqrt(v_b) + epsilon)) * db

        # Accuracy
        accuracy = np.mean((y_pred > 0.5) == y)

        losses.append(loss)
        accuracies.append(accuracy)
        epoch_list.append(i)

        if i % 100 == 0:
            print(f"Epoch {i} | Loss: {loss:.4f} | Accuracy: {accuracy:.4f}")

    return weights, bias, losses, accuracies, epoch_list
```

```python
weights, bias, losses, accuracies, epochs = RMSprop(X, y, epochs=200, lr=0.01, beta=0.9)
```

```python
plt.plot(epochs, accuracies, label='Accuracy')
plt.plot(epochs, losses, label='Loss')

plt.xlabel('Epochs')
plt.ylabel('Value')
plt.title('RMSprop: Loss vs Accuracy')
plt.legend()
plt.show()
```
